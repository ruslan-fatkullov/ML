```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, accuracy_score
from sklearn.linear_model import Perceptron, LinearRegression
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler, MaxAbsScaler, Normalizer
from sklearn.preprocessing import MinMaxScaler, RobustScaler, QuantileTransformer

%config InlineBackend.figure_format = 'svg'
pd.set_option("display.precision", 2)

test = pd.read_csv('Lab4Data/pendigits.tes', header=None)
train = pd.read_csv('Lab4Data/pendigits.tra', header=None)
x_train = train.drop([16], axis=1)
x_test = test.drop([16], axis=1)
y_train = train[16]
y_test = test[16]
```

```python

alphas = [0.0001, 0.001, 0.01, 0.1, 1]
penaltys = [None, 'l1', 'l2', 'elasticnet']

scores = []
errors = []
for index, pen in enumerate(penaltys):
    scores.append([])
    errors.append([])
    for alp in alphas:
        model = Perceptron(penalty=pen,verbose=True, alpha=alp)
        model.fit(x_train, y_train)
        scores[index].append(accuracy_score(y_test, model.predict(x_test)))        
        errors[index].append(accuracy_score(y_train, model.predict(x_train)))
```

```python
plt.rcParams["figure.figsize"] = (10,5)

penaltysLabel = ['None', 'l1', 'l2', 'elasticnet']

for (eror, score, title) in zip(errors, scores, penaltysLabel):
    plt.plot(alphas, score, label=title)
    plt.plot(alphas, eror, label=title)
plt.grid()
plt.legend()
plt.show()
```
